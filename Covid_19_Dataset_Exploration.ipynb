#**1.Importing Libraries:**
1. **matplotlib.pyplot**: This library provides a MATLAB-like plotting framework. It is commonly used for creating static, interactive, and animated visualizations in Python.
2. **seaborn**: Seaborn is a statistical data visualization library based on Matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
3. **pandas**: Pandas is a powerful data analysis and manipulation library. It provides data structures and functions to work with structured data, such as CSV files and DataFrames.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#**2.Loading the Dataset:**
pd.read_csv(): This function is used to read a CSV (Comma-Separated Values) file into a DataFrame. The file "dcov (1).csv" is read into the DataFrame df.


```python
# Load the dataset
df = pd.read_csv("dcov (1).csv")
```

#**3.Displaying Data:**
df.head(): This method returns the first n rows of the DataFrame. By default, n is 5. It is useful for quickly inspecting the structure and contents of the dataset.

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(df.head())
```

#**4.Summary Statistics of Numerical Features:**
df.describe(): This method generates descriptive statistics of the numerical columns in the DataFrame, such as count, mean, standard deviation, minimum, maximum, and quartiles.

```python

# Summary statistics of numerical features
print("\nSummary statistics of numerical features:")
print(df.describe())
```

#**5.Correlation Analysis:**
df.corr(): This method calculates the pairwise correlation of numerical columns in the DataFrame. It returns a correlation matrix where each cell represents the correlation coefficient between two variables.

```python
# Correlation matrix
print("\nCorrelation matrix:")
corr_matrix = df.corr()
print(corr_matrix)
```

#**6.Heatmap of Correlation Matrix:**
* sns.heatmap(): This function is used to
plot a heatmap of the correlation matrix. It visualizes the correlation coefficients between pairs of variables using color gradients.
* Parameters like annot, cmap, fmt, and annot_kws are used to customize the appearance of the heatmap.

```python

# Heatmap of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title("Correlation Matrix Heatmap")
plt.show()
```

#**7.Distribution of Age:**
* sns.histplot(): This function is used to plot a histogram of the specified column ('Age' in this case) from the DataFrame. It shows the distribution of values and can optionally include a kernel density estimation (KDE) curve.
* Parameters like bins, kde, and color are used to customize the appearance of the histogram.




```python

# Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

#**8.Relationship between Age and Heart Rate:**
* sns.scatterplot(): This function is used to create a scatter plot of two variables ('Age' and 'heart_rate' in this case) from the DataFrame. It shows the relationship between the variables.
* Parameters like x, y, data, and color are used to specify the variables and customize the appearance of the scatter plot.

```python
# Relationship between Age and Heart Rate
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='heart_rate', data=df, color='salmon')  # Adjusted column name to 'heart_rate'
plt.title("Relationship between Age and Heart Rate")
plt.xlabel("Age")
plt.ylabel("Heart Rate")
plt.show()
```

#**9.Boxplot of Heart Rate by Class:**
* sns.boxplot(): This function is used to create a boxplot to visualize the distribution of a numerical variable ('heart_rate') across different categories ('class' in this case).
* Parameters like x, y, data, and palette are used to specify the variables and customize the appearance of the boxplot.

```python
# Boxplot of Heart Rate by Class
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='heart_rate', data=df, palette='Set2')  # Adjusted column name to 'class'
plt.title("Boxplot of Heart Rate by Class")
plt.xlabel("Class")
plt.ylabel("Heart Rate")
plt.show()
```

#**10.Pairplot for Selected Features:**
* sns.pairplot(): This function creates a grid of pairwise scatter plots for the selected features. It helps to visualize relationships between multiple variables simultaneously.

```python
# Pairplot for selected features
selected_features = ['Age', 'Duration', 'Incoming_calls', 'Active_time']  # Updated selected features
sns.pairplot(df[selected_features])
plt.show()
```

#**11.Visualize Feature Distribution by Class:**
* This loop iterates over each feature in the DataFrame (except the last column, which is assumed to be the target variable 'class') and plots histograms for each feature, colored by class label. It helps to visualize how feature distributions vary across different classes.

```python
for feature in df.columns[:-1]:
    plt.figure(figsize=(8, 6))
    sns.histplot(x=feature, hue='class', data=df, kde=True)
    plt.title(f'Distribution of {feature} by Class')
    plt.show()
```

#**12.Box Plots for Selected Features:**
* This loop iterates over selected features and creates box plots to compare their distributions between different classes.

```python
selected_features = ['Duration', 'Incoming_calls', 'outgoing_calls', 'incoming_messages', 'Active_time', 'posts', 'nervous', 'sleephr', 'walkinghr', 'exercisehr', 'idlehr', 'heart_rate']
for feature in selected_features:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='class', y=feature, data=df)
    plt.title(f'Box Plot of {feature} by Class')
    plt.show()
```
